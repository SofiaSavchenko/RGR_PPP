```python
import pandas as pd
from scipy import stats

df_character = df['Price'][:200]
alpha = 1e-3

a, b = stats.normaltest(df_character)
if b < alpha: 
  print("The hypothesis can be rejected")
else:
  print("The hypothesis cannot be rejected")
```
